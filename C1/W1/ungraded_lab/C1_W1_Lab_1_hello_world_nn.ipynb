```python
import altair as alt
import numpy as np
import tensorflow as tf
from keras import layers
from tensorflow import keras
import pandas as pd

print(tf.__version__)
```

```python
input = layers.Input((1,))
mul = layers.Multiply()([input, input])
mul_dense = layers.Dense(units=1)(mul)
dense = layers.Dense(units=1)(input)
out = layers.Add()([mul_dense, dense])

optimizer = keras.optimizers.SGD(
    learning_rate=0.001,
    momentum=0.9,
)
model = keras.models.Model(inputs=[input], outputs=out)
model.compile(optimizer=optimizer, loss='mean_squared_error')
# model.summary()
```

```python
def f(x):
    return 2 * x ** 2 + 3 * x - 5


xs = np.arange(-2, 5, dtype=float)
ys = np.fromfunction(lambda i: f(xs[i]), xs.shape, dtype=int)

print(xs)
print(ys)
```

```python
model.fit(xs, ys, epochs=500, batch_size=2)
```

```python
xs_graph = np.linspace(-10, 10, 500, dtype=float)
ys_expected = np.fromfunction(lambda i: f(xs_graph[i]), xs_graph.shape, dtype=int)
ys_actual = np.fromfunction(lambda i: model.predict([xs_graph[i]]).squeeze(), xs_graph.shape, dtype=int)

# print(xs_graph)
# print(ys_expected)
# print(ys_actual)
```

```python
df = pd.DataFrame(list(zip(xs_graph, ys_expected, ys_actual)), columns=["x", "y_expected", "y_actual"]).melt("x")
alt.Chart(df).mark_line().encode(
    x='x',
    y='value',
    color='variable'
).interactive()
```
